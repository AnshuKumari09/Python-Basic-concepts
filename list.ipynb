```python
def myDecor(func):
    def wrapper():
        print("functn started")
        func()
        print("functn ended")
    return wrapper

@myDecor
def greet():
    print("good morning")

greet()
```

```python
# login check decor
is_logged_in=True
def login_required(func):
    def wrapper():
        if is_logged_in:
            func()
        else:
            print("plz login first")
        
    return wrapper


@login_required
def dashboard():
    print("welcome user")
```

```python
import time
def time_decor(func):
    def wrapper():
        start=time.time()
        func()
        end=time.time()
        diff=end-start
        print(diff)
    return wrapper
@time_decor
def task():
    time.sleep(2)

task()
```

```python
class myrange:
    def __init__(self,start,end):
        self.value=start
        self.end=end
    def __iter__(self):
        return self
    def __next__(self):
        if self.value>self.end:
            raise StopIteration
        current=self.value
        self.value=self.value+1
        return current
c=myrange(3,9)
for x in c:
    print(x)
```

```python
import time


def time_it(func):

    def wrapper():

        start = time.time()

        func()

        end = time.time()

        print("Time taken:", end - start)

    return wrapper


@time_it
def factorial():

    n = 5

    answer = 1

    for i in range(1, n+1):
        answer = answer * i

    print(answer)


factorial()
```

```python
# nestec decorator
def uppercase(func):
    def wrapper():
        text=func()
        
        return text.upper()
    return wrapper

def exclaim(func):
    def wrapper():
        text=func()
        return text+"!"
    return wrapper

@exclaim
@uppercase
def hello():
    return "hello"

print(hello())
```
