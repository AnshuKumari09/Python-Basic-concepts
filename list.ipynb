```python
# normal loop
numbers=[]
for i in range(1,10):
    numbers.append(i)
print(numbers)
```

```python
# list comprehension
numbers=[i for i in range(5)]
print(numbers)
```

```python
# square numbers
square_num=[x*x for x in range(5)]
print(square_num)
```

```python
# even numbers
even_nums=[x for x in range (5) if x%2==0]
print(even_nums)
```

```python
# do loops
pairs=[]
for i in range(3):
    for j in range(2):
        pairs.append((i,j))

pairs
```

```python
pairs2=[(i,j) for i in range(3) for j in range(2)]
print(pairs2)
```

```python
result=["Even" if x%2==0 else "odd" for x in range(5)]
result
```

```python
res=["small" if x<=2 else "big" for x in range(6)]
res
```

```python
#filter\
nums=1,4,6,7,9,0
f1=list(filter(lambda x:x%2==0,nums))
f1
```

```python
nums=[2,-9,6,3,-7]
f2=list(filter(lambda x:x>=0,nums))
f2
```
